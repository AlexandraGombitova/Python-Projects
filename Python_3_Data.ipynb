## III. üêº DataFrame

```python
import pandas as pd
import numpy as np

index_a = [1, 2, 3, 4, 5, 6, 7, 8, 9]
columns_a = ["First", "Second", "Third"]
dataframe_a = pd.DataFrame(np.random.randn(9,3), index = index_a, columns = columns_a)
print("\nFirst Table")
print(dataframe_a)



index_b = ["a", "b", "d", "e", "f", "h", "i", "j", "k"]
columns_b = ["First", "Second", "Third"]
dataframe_b = pd.DataFrame(np.random.randn(9,3), index = index_b, columns = columns_b)
dataframe_b = dataframe_b.reindex(["a", "b", "c", "d", "e", "f", "h", "i","j"])
print("\nSecond Table")
print(dataframe_b)
print("\nNull Table")
print(dataframe_b.isnull())
print("\nFixed Table")
print(dataframe_b.fillna(0))
```

## IV. üìä Histogram 

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np

mi, sigma = 100, 20
x = mi + sigma * np.random.randn(1000)
n, bins, patches = plt.hist(x, 20, density = 1, facecolor = "m",)
plt.xlabel("IQ")
plt.ylabel("Propability")
plt.title("IQ Histogram")
plt.grid()
plt.show
```
