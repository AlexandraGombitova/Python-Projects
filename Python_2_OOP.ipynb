```python
import math

class Shape:
    size = float()
    color = str()
    perimeter = float()
    area = float()

    def __init__(self, size):
        self.size = size

    def __init__(self, size, color):
        self.size = size
        self.color = color

class Square(Shape):
    def calculate_the_area_of_square(self):
        return math.pow(self.size, 2)

    def calculate_the_perimeter_of_square(self):
        return 4 * self.size


class Circle(Shape):
    def calculate_the_area_of_circle(self):
        return math.pi * math.pow(self.size, 2)

    def calculate_the_perimeter_of_circle(self):
        return 2 * math.pi * self.size    


shape_1 = Shape(4.6 , "Black")
square_1 = Square(5.0, "Red")
print("Square:")
print(square_1.calculate_the_area_of_square())
print(square_1.calculate_the_perimeter_of_square())
print("\nCircle:")
circle_1 = Circle(7.0, "Blue")
print(circle_1.calculate_the_area_of_circle())
print(circle_1.calculate_the_perimeter_of_circle())
```

## IV. Multiple Ineritance

```python
class Father:
    def do_A(self):
        print("A")
    def do_something(self):
        print("Father -  something")

class Mother:
    def do_B(self):
        print("B")
    def do_something(self):
        print("Mother -  something")

class Son(Father, Mother):
    def do_C(self):
        print("C")
    def do_something(self):
        print("Son -  something")
        super().do_something()
        Mother.do_something(self)
       

print("Father:")
peter = Father()
peter.do_A()
peter.do_something()

print("\nMother:")
alexa = Mother()
alexa.do_B()
alexa.do_something()

print("\nSon:")
miki = Son()
miki.do_C()
miki.do_A()
miki.do_B()
miki.do_something()
```
