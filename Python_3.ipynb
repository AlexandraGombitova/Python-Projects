## 3. If

```python
age = abs(int(input("How old are you ? ")))
print("You are" , age, "years old...")

if age >= 18:
    print("You are eligible to drive a car")
else:
    print("You are not eligible to drive a car")
    
```

```python
age = abs(int(input("How old are you ? ")))
print("You are" , age, "years old...")

if age < 18:
    print("You are not eligible to drive a car")

elif age < 90:
    print("You are eligible to drive a car")

else:
    print("You are not eligible to drive a car")
    
```

```python
age = abs(int(input("How old are you ? ")))
print("You are" , age, "years old...")

if age >= 18 and age < 90:
    print("You are eligible to drive a car")

else:
    print("You are not eligible to drive a car")
```

## 4. Logical operations

```python
#and
print("\nAnd")
print(True and True)
print(True and False)

#or
print("\nOr")
print(True or True)
print(True or False)

#not
print("\nNot")
print(not(True or False))
print(not(True and False))

#= assignement
print("\n=")
a = 65
print(a)

#== comparism
print("\n==")
print(a==a)
```

## 5. For

```python
list_of_users = ["Iva", "Peter", "Alex", "Jana", "Dona", "Miki"]
print("Content of list --> ", list_of_users)
print("Count of aspects in list --> ", len(list_of_users))
print("First user --> ", list_of_users [0])
print("Last user --> ", list_of_users [-1])

print("\nRange:")
numbers = range(9)
print(list(numbers))
numbers1 = range(1,19)
print(list(numbers1))
numbers2 = range(1,11,2)
print(list(numbers2))


print("\nIn:")
list_of_users = ["Iva", "Peter", "Alex", "Jana", "Dona", "Miki"]
print("Miki" in list_of_users)
```

```python
for i in range(5):
    print( i, "Hi Alexa")
    
```

```python
list_of_users = ["Iva", "Peter", "Alex", "Jana", "Dona", "Miki"]
print("List of users:")
for user in list_of_users:
    print(user)
```

```python
result = 0
for i in range(1, 4):
    result = result + i

print("Result will be: ")
print(result)
```

```python
list_of_users = ["Iva", "Peter", "Alex", "Jana", "Dona", "Miki"]
print("Reversed list:")
for user in reversed(list_of_users):
    print(user)
```

```python
list_of_users = ["Iva", "Peter", "Alex", "Jana", "Dona", "Miki"]
print("Numbered list:")
for i, user in enumerate(list_of_users):
    print(i+1, "-->", user)
```

## 6. While 

```python
i = 1
while i < 10:
    print(i, "Hi Alex!")
    i += 1

print("We are at the end....")
```

```python
year = 2025
while True:
    print("I want to live in the year", year)
    year += 1

    if year >= 2030:
        print("The end")
        break
    
   
```
