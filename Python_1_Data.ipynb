## II. Matplotlib

```python
try:
    from matplotlib import pyplot as plt
    import numpy as np
    import pandas as pd
    import math
    
    x = np.arange(0, 5);
    print(x)
    y = x ** 2;

    plt.title("Line chart --> x, y")
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.plot(x, y,)


    plt.annotate(xy = [1, 2], text = "Point_1")
    plt.annotate(xy = [2, 4], text = "Point_2")
    
except:
    print("Somthing failed....")



finally:
    print("In the Finally...")

```
