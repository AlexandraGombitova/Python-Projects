## 1. Files

```python
import datetime

today = datetime.datetime.now()
file = open("logy.txt", "w")
message = (str(today) + "-- is a nice weather")
file.write(message)
print("Written in the file....")
file.close()
```

```python
import datetime

today = datetime.datetime.now()
file = open("logy.txt", "a")
message = (str(today) + "-- is a nice weather")
file.write(message)
print("Written in the file....")
file.close()
```

```python
file = open("logy.txt", "r")
data = file.read()
print("Content of the file --> ", data)
```

## 2. Dict and Sets

```python
family = {"Mom" : "Jana", "Dad" : "Ivan", "Sister" : "Jaja" ,"Brother": "Momo"}
print("Content of family:" , family ,"Type -->", type(family))
for k in family:
    print("Key is -->", k, "and has a value -->", family[k])
```

```python
users = ["Pete","Miki", "Iva", "Momo", "Jana"]
roles = ["Gamer", "Dancer", "Driver","Criminallist","Victim"]
dictonary_of_friends = dict(zip(users, roles))
print(dictonary_of_friends)
```

```python
set_of_products = {"phone", "tablet", "notebook", "usb", "disk"}
print("Content of set of products:" , set_of_products ,"Type -->", type(set_of_products))
```
