## II. Class

```python
class Dog:
    """ This is our first class named Dog
    """
    breed = " "
    name = str()
    is_lazy = bool()
    age = int()
    height = float()
    gender = list()
    size = tuple()


    
    def barking(self):
        print("Haf! Haf!")
    
    def sleeping(self):
        print("Dog is sleeping...")
    
    def print_info(self):
        print("What is dog´s name? ---> ", self.name)
        print("How old is the dog? ---> ", self.age)
        print("Is the dog lazy? ---> ", self.is_lazy)

    


    
miki = Dog()
miki.name = "Miki"
miki.age = 2
miki.breed = "French Bulldog"
miki.is_lazy = True
miki.height = 45.78
miki.gender= ("Male")
miki.size = (3, 45, 67)

print(miki.name)
print(miki.age)
print(miki.breed)
print(miki.is_lazy)
print(miki.height)
miki.print_info()
```
