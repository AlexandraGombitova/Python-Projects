## I. âž— Def

```python
import math
def count_volume_sphere(r = 1):
    """
    Count the volume of sphere
    """
    r = float(r)
    return 4/3 * math.pi * math.pow(r, 3)


count_volume_sphere(10)
```

```python
def create_email(m, p, domain = "abc.sk"):
    email =  m.strip() + "." + p.strip() + "@" + domain
    email = email.lower()
    return email


create_email("Alex  ", "Gomb  ")
```

```python
def pow_q(a):
    return lambda q: q**a

second_pow = pow_q(2)
print(second_pow(10))
```

## IV. ðŸŒŒ SciPy 

```python
import matplotlib.pyplot as plt
from scipy import datasets # This is the correct module for 'face'

# 1. Load the racoon image data (now works because 'pooch' is installed)
face = datasets.face()

# 2. Print information about the image array
print(f"Data type: {type(face)}")
print(f"Shape (Height, Width, Color Channels): {face.shape}")
print(f"Maximum intensity value: {face.max()}")

# 3. Plot the image
plt.figure()
plt.axis("off")      # Turn off axis labels and ticks
plt.gray()           # Apply a grayscale colormap
plt.imshow(face)     # Display the image array
plt.title("The SciPy Raccoon (in Grayscale)")
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import datasets
import numpy as np

def shades(img, percentage):
    picture_shade = img + (np.ones(img.shape) - img) * percentage
    return picture_shade


python_logo = plt.imread("logo.png")
plt.axis("off")
plt.imshow(python_logo)
plt.show()
print(python_logo.shape)

plt.axis("off")
python_logo2 = shades(python_logo, 0.9)
plt.imshow(python_logo2)
plt.show()
```
