## 2. Data Types in Python

```python
# string/str

client_name = ("Alexandra Gombitova")
product_name = ("ARTEMIS")
print("What type is client_name ? - " , type(client_name))

client_namr = input("What is your name? - ")
print("Hi!" , client_name)
```

```python
# integer/int

client_age = 21
print("What type is client_age ?" , type(client_age))
```

```python
#float 

product_price = 21.34
print("What type is product_price ?" , type(product_price))
print(product_price)
```

```python
#Bool

is_married = True
print("What type is is_married ?" , type(is_married))
print(is_married)
```

```python
#None

is_registered = None
print("What type is is_registered ?" , type(is_registered))
print(is_registered)
```

```python
#Complex

a = 3 + 3j
print("What type is a ?" , type(a))
print(a)
```

```python
#Bytes

a = b"HI"
print("What type is a ? " , type(a))
print(a)
```

```python
#memoryview

b = memoryview(bytes(2))
print(b)
print("What type is b ? " , type(b))
```
